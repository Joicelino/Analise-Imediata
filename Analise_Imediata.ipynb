```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patheffects import withSimplePatchShadow

# Dados
data = {
    "Espcie": [
        "Vachellia tortilis", 
        "Brachystegia spiciformis", 
        "Parthenium hysterophorus", 
        "Argemone mexicana ", 
        "Ficus spp."
    ],
    "Mdia_TH": [8.29, 7.95, 10.97, 8.40, 10.65],
    "SD_TH": [0.19, 0.16, 0.04, 0.20, 0.12]
}

df = pd.DataFrame(data)

# Altura mxima para eixo Y
y_max = max(df["Mdia_TH"] + df["SD_TH"]) + 0.8  

# Criar figura
fig, ax = plt.subplots(figsize=(10,6))

# Barras com sombra discreta
bars = ax.bar(
    df["Espcie"], df["Mdia_TH"], 
    yerr=df["SD_TH"], capsize=6, 
    color="#4a90e2", edgecolor="black", alpha=0.85, zorder=3
)

shadow_effect = withSimplePatchShadow(offset=(2,-2), shadow_rgbFace='grey', alpha=0.25)
for bar in bars:
    bar.set_path_effects([shadow_effect])

# Personalizao do grfico
ax.set_ylim(0, y_max)
ax.set_title("Teor de Humidade (TH) das Espcies", fontsize=16, fontweight='bold', pad=15)
ax.set_ylabel("Teor de Humidade (%)", fontsize=14)
plt.xticks(rotation=20, fontsize=12, ha='right')  # alinha  direita se quiser
ax.tick_params(axis='x', pad=10)  # aumenta a distncia entre rtulo e eixo X

# Grade suave apenas no eixo Y
ax.grid(axis='y', linestyle='--', alpha=0.4, zorder=0)

# Salvar o grfico em arquivo
# Afasta os rtulos das espcies do eixo X

plt.tight_layout()  # Ajusta o layout para que nada fique cortado
plt.savefig("grafico_TH_especies.png", dpi=300)  # Salva em PNG com alta resoluo
plt.savefig("grafico_TH_especies.pdf")  # Opcional: salvar tambm em PDF

# Remover bordas desnecessrias
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.8)
ax.spines['bottom'].set_linewidth(0.8)
plt.tight_layout()
plt.show()
```
